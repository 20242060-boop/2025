```python
class Bus:

    def __init__(self, passengers=None):

        if passengers is None:

            self.passengers = []
        else:

            self.passengers = list(passengers)


    def pick(self, name):
        self.passengers.append(name)


    def drop(self, name):

        self.passengers.remove(name)
```

```python
import copy
bus1 = Bus(['Alice', 'Bill', 'Claire', 'David'])
bus2 = copy.copy(bus1)
bus3 = copy.deepcopy(bus1)
id(bus1), id(bus2), id(bus3)
```

```python
bus1.drop('Bill')
bus1.passengers, bus2.passengers, bus3.passengers
```

```python
id(bus1.passengers), id(bus2.passengers), id(bus3.passengers)
```

```python
class HauntedBus:
    """유령 승객이 출몰하는 버스 모델"""


    def __init__(self, passengers=[]):

        self.passengers = passengers

    def pick(self, name):
        self.passengers.append(name)

    def drop(self, name):
        self.passengers.remove(name)
```

```python
bus1 = HauntedBus(['Alice', 'Bill'])
print(bus1.passengers)
```

```python
bus1.pick('Charlie')
bus1.drop('Alice')
print(bus1.passengers)
```

```python
bus2 = HauntedBus()
bus2.pick('Carrie')
print(bus2.passengers)
```

```python
bus3 = HauntedBus()
print(bus3.passengers)
```

```python
bus3.pick('Dave')
print(bus2.passengers)
```

```python
bus2.passengers is bus3.passengers
```

```python
class TwilightBus:
    """승객이 사라지거나 안 타는 버스 모델"""

    def __init__(self, passengers=None):
        if passengers is None:
            self.passengers = []
        else:

            self.passengers = list(passengers)

    def pick(self, name):
        self.passengers.append(name)

    def drop(self, name):
        self.passengers.remove(name)
```

```python
basketball_team = ['Sue', 'Tina', 'Maya', 'Diana', 'Pat']


bus = TwilightBus(basketball_team)


bus.drop('Tina')
bus.drop('Pat')


basketball_team
```

```python
class TwilightBus:
    """승객이 사라지게 만드는 버스 모델"""

    def __init__(self, passengers=None):
        if passengers is None:
            self.passengers = []
        else:

            self.passengers = list(passengers)

    def pick(self, name):
        self.passengers.append(name)

    def drop(self, name):
        self.passengers.remove(name)
```

```python
basketball_team = ['Sue', 'Tina', 'Maya', 'Diana', 'Pat']


bus = TwilightBus(basketball_team)


bus.drop('Tina')
bus.drop('Pat')


print(basketball_team)
```
